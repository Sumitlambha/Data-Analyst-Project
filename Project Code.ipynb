```python
#changing data type of value attributes from string to float
for i in range(len(new_file['Value'])):
    new_val = ""
    for char in new_file.loc[i,"Value"]:
        if char.isdigit() or char == '.':
            new_val+=char
    new_file.loc[i,"Value"] = float(new_val)



                                     
```

```python
#Data Modelling
from sklearn import preprocessing
#limiting the data frames to contain the attributes required for prediction
new_file = new_file[['Overall','Potential','Special','International Reputation','BallControl','Skill Moves','Reactions','Value']]
#drop row with all null values 
new_file = new_file.dropna()
```

```python
#Printing the limited dataframe 
new_file.head()
```

```python
#Encoding the value attributes for prediction
le = preprocessing.LabelEncoder()
new_file["Value_label"] = le.fit_transform(new_file["Value"])
```

```python
#Printing the data frames after addition of value label
new_file.head()
```

```python
#Collecting the attributes used for pridiction in X and attributes to be pridicted in Y
X = new_file.iloc[:,[0,1,2,3,4,6,8]]
Y = new_file.iloc[:,5]
new_file.iloc[:,[5]]
```

```python
#Splitting the set x and y in ratio of 50% for training and 50% for testing
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(X, Y, test_size=0.5)
```

```python
#Printing few rows of trainning sets
print(X_train.head(10))
print(Y_train.head(10))
```

```python
#Appling KNN with K=100
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 200, metric='minkowski', p=2)
classifier.fit(X_train, Y_train)
Y_pred= classifier.predict(X_test)
```

```python
#Printing the result of confusion metrix and classification report
from sklearn import metrics
from sklearn.metrics import classification_report
print classification_report(Y_test,Y_pred)
print(metrics.confusion_matrix(Y_test, Y_pred))
```
